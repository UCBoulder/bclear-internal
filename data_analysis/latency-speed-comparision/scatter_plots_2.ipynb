```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import json
import io
```

## Idle latency vs Average download speed (Jan 2011 - Jul 2023)

```python
latency = pd.read_csv("data_rtt_max/Average of Idle latency.csv")
speed = pd.read_csv("data_rtt_max/Average download speed.csv")
merge_df = pd.merge(speed, latency, how="left", on="dtime per day")
merge_df.tail()
```

```python
# Calculate fit line
x = merge_df["Average speed (Mbps)"]
y = merge_df["Average latency (ms)"]
# idx = np.isfinite(x) & np.isfinite(y)
# up_trend = np.polyfit(x[idx], y[idx], 3)
# up_fit = np.poly1d(up_trend)

plt.figure(figsize=(12,3))
sns.set_theme(style="whitegrid")
plt.scatter(x, y, s=3.5)
# plt.plot(x, up_fit(x), "r--", label="Upstream Trend")
plt.xlabel('Download speed (Mbps)')
plt.ylabel('Idle latency rtt_max (ms)')
plt.show()
```
