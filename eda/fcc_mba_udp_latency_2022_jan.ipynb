```python
# Data cleansing PASS 2 - remove outliers
# Apply Tukey's rule to filter
```

```python
# Calculate Interqaurtile Range
Q1 = latency['rtt_max'].quantile(0.25)
Q3 = latency['rtt_max'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Calulate lower boundry of filter
lower_lim = Q1 - 1.5 * IQR
# Calulate upper boundry of filter
upper_lim = Q3 + 1.5 * IQR
```

```python
# Apply filter to rtt_max vector
outliers_low = (latency['rtt_max'] < lower_lim)
outliers_up = (latency['rtt_max'] > upper_lim)
```

```python
# Outliers filtered from rtt_max?
num_outliers = len(latency['rtt_max'][(outliers_low | outliers_up)])
latency['rtt_max'][(outliers_low | outliers_up)]
```

```python
# Impact of outlier filter?
num_passed = len(latency['rtt_max'][~(outliers_low | outliers_up)])
latency['rtt_max'][~(outliers_low | outliers_up)]
```

```python
# Percentage of outliers Tukey Filtered?
```

```python
100 * (num_outliers/num_rows) 
```

```python
# Strip outliers from rtt_max
```

```python
latency_rtt_max_no_outliers = latency['rtt_max'][~(outliers_low | outliers_up)]
latency_rtt_max_no_outliers=latency_rtt_max_no_outliers[latency_rtt_max_no_outliers!=0]
```
